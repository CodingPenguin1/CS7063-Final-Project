```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results/ga_cifar_10_epochs_20_generations.csv')
# df = pd.read_csv('results/ga_mnist_10_epochs_5_generations.csv')
df
```

```python
plt.plot(df['Generation'], df['Champion Fitness'], label='Max fit')
plt.plot(df['Generation'], df['Average Fitness'], label='Avg fit')
plt.plot(df['Generation'], df['Minimum Fitness'], label='Min fit')
plt.legend(loc='upper right')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Fitness')
plt.show()
```

```python
plt.plot(df['Generation'], df['Champion Accuracy'], label='Max acc')
plt.plot(df['Generation'], df['Average Accuracy'], label='Avg acc')
plt.plot(df['Generation'], df['Minimum Accuracy'], label='Min acc')
plt.legend(loc='upper right')
plt.xlabel('Generation')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.show()
```

```python
plt.plot(df['Generation'], df['Average Capacity'], label='Avg cap')
plt.plot(df['Generation'], df['Champion Capacity'], label='Min cap')
if 'Maximum Capacity' in df.columns:
    plt.plot(df['Generation'], df['Maximum Capacity'], label='Max cap')
plt.legend(loc='upper right')
plt.xlabel('Generation')
plt.ylabel('Capacity')
plt.title('Capacity')
plt.show()
```
